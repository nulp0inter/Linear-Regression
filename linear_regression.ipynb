```python
import pandas as pd
```

```python
df_1 = pd.DataFrame({ 
    'x':[1,2,3,4,5],
    'y':[5,7,9,11,13] #y = 2*x + 3
})

df_2 = df_1.copy(deep=True)

df_1
```
