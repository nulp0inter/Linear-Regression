```python
#https://archive.ics.uci.edu/dataset/9/auto+mpg

import pandas as pd
```

```python
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
```

```python
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']

df = pd.read_csv(url, delim_whitespace=True, names=column_names, na_values='?', comment='\t')

df
```

```python
df.isna().sum()
```

```python
df['horsepower'].median()
```

```python
df['horsepower'].mean()
```

```python
df['horsepower'].fillna(df['horsepower'].median(),inplace=True)
```

```python
df.isna().sum()
```

```python
X = df.drop(columns=['mpg','car_name'])
```

```python
X
```

```python
y = pd.DataFrame()
y['mpg'] = df['mpg'].copy(deep=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=13)
```

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler()

X_train = Scaler.fit_transform(X_train[X_train.columns])
X_test = Scaler.fit_transform(X_test[X_test.columns])

#no not scale y_train & y_test

'''
Points to note for linear regression

    * Split the dataset into X (independent varaibles used for predicting y) and y (dependent variable which will 
    be predicted) from oiginal dataframe 
    
    * After splitting further split into X_train, X_test, y_train, y_test then scale X_train & X_test with their
    respective minimun & maximum value OR mean & standard deviation. Here, do not scale y_train & y_test. test
    
    * The reason we dont scale y_train & y_test because if you scale y_train after 
                                    
                                    y_pred = model.predict(X_test)
      
      returns y_pred will be in range 0 - 1. Here, ideal MAX value in y_pred = 1 and ideal MIN value in y_pred = 0
      but this never happens. MAX value in y_pred() can bee less than 1.02 or 0.86 and MIN value in y_pred() can be
      0.21 or -0.02. Now, we might rescale it back by
      
      y_pred = y_pred * (y_test.max() - y_test.min()) + y_test.min() 

      this will never work as y_test.max() ≠ y_test.max() & y_test.min() ≠ y_test.min()
      
      Also, rescaling y_pred to 0-1 range and rescaling it to y_test.max() - y_test.min() is not best solution either.

     * After 
     
                                    y_pred = model.predict(X_test)
                                    
      y_pred is in numpy so change it to dataframe with column name and index as of y_test by
      
                          y_pred = pd.DataFrame(y_pred, index=y_test.index, columns=y_test.columns)
                          
      and perform model evaluation like MSE, MAE, R2 etc 
      
      
'''
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred[:15]) #here y_pred is numpy array and y_test is dataframe
print()
print(y_test[:15])
```

```python
y_pred = pd.DataFrame(y_pred, index=y_test.index, columns=y_test.columns)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.metrics import r2_score 

r2_score(y_test,y_pred)
```

```python
model.score(X_test,y_test)
```
